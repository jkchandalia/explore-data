```python
with open("temp_data") as file:
    N = int(file.readline())
    print(N)
    columns = file.readline().split()
    test = []
    train = []
    for _ in range(N):
        line = file.readline()
        if 'Missing' in line:
            test.append(line.split())
        yyyy, month, tmax, tmin = line.split()
        if 'Missing' in tmax:
            tmax = np.nan
        if 'Missing' in tmin:
            tmin = np.nan
        train.append([yyyy, month, tmax, tmin])
        
    
```

```python
#Plot boxcox for diff lambdas
from matplotlib import pyplot as plt
x=np.array(range(0,100))+.1
x
y={}

def boxcox(arr, lmbda):
    if lmbda == 0:
        return np.log(arr)
    else:
        return (arr**lmbda - 1) / lmbda
lmbda_vals = [-10,0,-1,-.1, .1, 1, 2, 3, 4, 5]
for i in range(len(lmbda_vals)):
    y[i] = boxcox(x, lmbda_vals[i])
    
for i in range(len(y.keys())):
    plt.plot(x,y[i], label=lmbda_vals[i])
plt.yscale("log")
plt.legend()
```
